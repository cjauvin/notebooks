## Ma compréhension de la PCA

Supposons que je définisse un dataset $X$ dans $R^2$, où les deux dimensions sont fortement corrélées (avec un peu de bruit ajouté pour faire bonne mesure!).

```python
n = 100
x = random.random(n)
noise = random.normal(scale=0.5, size=n)
y = x * 3 + noise
X = vstack((x, y))
scatter(x, y);
```

La forte corrélation entre les deux dimensions est aussi exprimée par le fait que les éléments non-diagonaux de la matrice de covariance $S$ sont $>> 0$.

```python
X_centered = X.T - mean(X, axis=1)
S = np.dot(X_centered.T, X_centered) / (n - 1)
assert(array_equal(S, cov(X))) # juste pour vérifier que je fais la même chose que Numpy
S
```

```python
X = random.multivariate_normal((3, 5), ([0.05,   0.2], 
                                        [0.2,   0.5]),   1000)
scatter(X[:, 0], X[:, 1])
cov(X.T)
```

Le but de la PCA est de trouver une matrice $P$ qui constitue une nouvelle base orthonormale vers laquelle les données $X$ seront projetées (à partir donc de la base orthogonale implicite $I$ dans laquelle $X$ avait été préalablement définie), donc  $P \cdot X = Y$. Toutefois, on ajoute la contrainte que la matrice de covariance de $Y$ devrait être diagonale, ce qui devrait découler du fait que les dimensions de $Y$ ont été maximalement décorrélées. La façon d'y parvenir est de définir $P$ en tant que les eigen-vecteurs de $S$ (ici s'arrête pas mal ma compréhension en fait, parce que je n'ai pas poussé mon étude jusqu'à comprendre pourquoi exactement).

```python
P = linalg.eig(S)[1]
Y = np.dot(P.T, X)
cov(Y)
```

Si on représente graphiquement $Y$, on voit que la projection vers $P$ a deux effets : (1) réaligner sur le plan les deux dimensions où la variance est maximale, et (2) décorréler les deux dimensions :

```python
scatter(Y.T[:, 0], Y.T[:, 1]);
```

Si on modifie l'expérience de manière à réduire le bruit à presque rien :

```python
x = random.random(n)
noise = random.normal(scale=0.0001, size=n)
y = x * 3 + noise
X = vstack((x, y))
scatter(x, y);
```
